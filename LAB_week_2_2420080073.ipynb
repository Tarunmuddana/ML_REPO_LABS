```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv(r"C:\Users\mudda\Downloads\Titanic-Dataset.csv")

print("titanic dataset")
print(df.shape[0], "number of rows")
print(df.shape[1],"number of columns")
print(df.isnull().sum(),"number of missing values ")
```

```python

age_medium = df["Age"].median()
df["Age"] = df["Age"].fillna(age_medium)

emabarked_mode = df["Embarked"].mode()[0]
df["Embarked"] = df["Embarked"].fillna(emabarked_mode)


df.drop(columns=["PassengerId","Name","Ticket","Cabin"],inplace=True)

print("Missing values after handling",df.isnull().sum())
```

```python
q1 = df["Fare"].quantile(0.25)
q3 = df["Fare"].quantile(0.75)

iqr = q3 - q1

lower_bound = q1 + 1.5*iqr
upper_bound = q3 - 1.5*iqr


df["Fare"] = np.where(df["Fare"] < lower_bound,df["Fare"],lower_bound)
df["Fare"] = np.where(df["Fare"] > upper_bound, df["Fare"], upper_bound)


 
```

```python
df = pd.get_dummies(df, columns=['Sex', 'Embarked'], drop_first=True)
le = LabelEncoder()
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])
```

```python
X = df.drop("Survived",axis=1)
Y = df["Survived"]

X_train , X_temp,Y_train,  Y_temp =train_test_split(X,Y,test_size=0.30,random_state=42,stratify=Y)
X_val , X_test , Y_val , Y_test = train_test_split(X_temp,Y_temp,test_size=0.50,random_state=42,stratify=Y_temp)

print("Datasset Split Shapes")
print("Training Set:" ,X_train.shape,Y_train.shape)
print("Validation set", X_val.shape,Y_val.shape)
print("Testing data",X_test.shape,Y_test.shape)
```
